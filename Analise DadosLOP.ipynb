<h1><center>Machine Learning</center></h1>:
<h2><center>Neural Network</center></h2>

```python
# Splitting the dataset into the Training set and Test set


# features: Mean =  ('qsub1','qsub2','qsub3') , Mean = ('igualACeml123', 'igualACeml45') , 
# Max = ('submeteu1','submeteu2','submeteu3','submeteu4','submeteu5','submeteu6','submeteu7','submeteu8')
# Mean = 'subListaExer23','subListaExer45','subListaExer67','subListaExer89'

df = pd.read_csv('https://raw.githubusercontent.com/ect-info/ml/master/dados/DataBaseLop.csv')

X = df.loc[:,['qsub1','qsub2','qsub3','igualACeml123','submeteu1','submeteu2','submeteu3','subListaExer23','subListaLab23'] ]
y = df.loc[:,'situacao']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Treinando a parte de encoder
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor

clf = MLPRegressor(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5,10,10,9,1), random_state=1)

clf.fit(X_train, y_train) 
print(clf)

y_pred = clf.predict(X_test)

#print(y_test)
y_pred = (y_pred > 0.5)
#print(y_pred)
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print("Matriz de Confusão:")
print(cm)
print("Taxa de acerto:")
print((cm[0,0]+cm[1,1])/len(y_test) )
print(len(y_test))
```
